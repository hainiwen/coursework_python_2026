# Exercise 2: Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(42)
```

## 2a: Loading and displaying data
_(mirrors food_facts.ipynb)_

```python
# Build a small food-facts style dataframe
food = pd.DataFrame({
    'product':  ['Oat milk', 'Cheddar', 'Chicken breast', 'White bread',
                 'Salmon', 'Spinach', 'Greek yogurt', 'Dark chocolate',
                 'Brown rice', 'Lentils'],
    'category': ['Dairy alt', 'Dairy', 'Meat', 'Bakery',
                 'Fish', 'Vegetable', 'Dairy', 'Confectionery',
                 'Grain', 'Legume'],
    'calories': [47, 402, 165, 265, 208, 23, 97, 546, 216, 116],
    'protein':  [1.0, 25.0, 31.0, 9.0, 20.0, 2.9, 10.0, 5.5, 4.5, 9.0],
    'fat':      [1.5, 33.0, 3.6, 3.2, 13.0, 0.4, 0.4, 31.0, 1.8, 0.4],
    'country':  ['SE', 'UK', 'US', 'FR', 'NO', 'IT', 'GR', 'BE', 'TH', 'FR'],
})

print('Shape:', food.shape)
print('\nColumn dtypes:')
print(food.dtypes)

food.head()
```

```python
# Basic descriptive statistics
food.describe()
```

```python
# Check for missing values
print('Missing values per column:')
print(food.isnull().sum())
```

```python
# Select individual columns and rows
print('All calories:')
print(food['calories'].values)

print('\nRow 2 (iloc):')
print(food.iloc[2])

print('\nProduct + protein columns:')
print(food[['product', 'protein']])
```

## 2b: Filtering and sorting
_(mirrors army.ipynb)_

```python
# Filter: high protein foods (> 10g)
high_protein = food[food['protein'] > 10]
print('High protein foods (>10g):')
print(high_protein[['product', 'protein', 'calories']])
```

```python
# Filter: low calorie AND low fat
light = food[(food['calories'] < 100) & (food['fat'] < 1.0)]
print('Low calorie (<100 kcal) AND low fat (<1g):')
print(light[['product', 'calories', 'fat']])
```

```python
# Filter: French products
french = food[food['country'] == 'FR']
print('French products:')
print(french[['product', 'category', 'calories']])
```

```python
# Sort by calories descending
print('Sorted by calories (highest first):')
print(food.sort_values('calories', ascending=False)[['product', 'calories']])
```

```python
# Sort by protein descending, calories ascending (multi-key sort)
print('Multi-key sort: protein desc, then calories asc:')
print(food.sort_values(['protein', 'calories'],
                       ascending=[False, True])[['product', 'protein', 'calories']])
```

```python
# Add a derived column: protein efficiency (protein per 100 kcal)
food['protein_per_100kcal'] = (food['protein'] / food['calories'] * 100).round(1)
print('Protein efficiency ranking:')
print(food.sort_values('protein_per_100kcal', ascending=False)
      [['product', 'protein_per_100kcal']])
```

## 2c: Grouping data
_(mirrors alcohol.ipynb)_

```python
# Mean calories per category
print('Mean calories by category:')
print(food.groupby('category')['calories'].mean().sort_values(ascending=False))
```

```python
# Multiple aggregations per group
summary = food.groupby('category').agg(
    n_products   = ('product',  'count'),
    mean_calories= ('calories', 'mean'),
    max_protein  = ('protein',  'max'),
    mean_fat     = ('fat',      'mean'),
).round(1)
print('Summary per category:')
summary
```

```python
# Count products per country
print('Products per country:')
print(food.groupby('country')['product'].count().sort_values(ascending=False))
```

```python
# Pivot table: mean calories & protein by category
pivot = food.pivot_table(
    values=['calories', 'protein'],
    index='category',
    aggfunc='mean'
).round(1)
print('Pivot table:')
pivot
```

```python
# Plot: mean calories by category
ax = (food.groupby('category')['calories']
          .mean()
          .sort_values()
          .plot(kind='barh', figsize=(8, 5), color='#2196F3',
                alpha=0.8, edgecolor='white'))
ax.set_xlabel('Mean calories (kcal / 100g)')
ax.set_title('Mean calories by food category')
ax.grid(axis='x', alpha=0.3, ls='--')
plt.tight_layout()
plt.savefig('food_categories.png', dpi=130, bbox_inches='tight')
plt.show()
```
